### 파일 선별하기 위해 사용

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2

def check(rgb_dir, color_dir, depth_dir):
    resize = 0.6
    img_rgb = cv2.resize(cv2.imread(rgb_dir),(0,0), fx = resize, fy = resize, interpolation = cv2.INTER_AREA)
    img_color = cv2.resize(cv2.imread(color_dir),(0,0), fx = resize, fy = resize, interpolation = cv2.INTER_AREA)
    img_align = cv2.addWeighted(img_rgb, 1, img_color, 0.3,0)
    blank = np.zeros((img_align.shape[0], img_align.shape[1]), dtype = np.uint8)
    blank = cv2.merge((blank,blank,blank))

    # img_rgb = cv2.imread('RGB_20220426_001.png')
    # img_color = cv2.imread('dist_20220426_001.png')

    compare1 = np.hstack((img_rgb, img_color))
    compare2 = np.hstack((img_align, blank))
    compare = np.vstack((compare1, compare2))

    winname = str(rgb_dir)
    cv2.namedWindow(winname)
    cv2.moveWindow(winname, 40,30)
    cv2.imshow(winname, compare)
    
    while True:
        k = cv2.waitKey()
        if k == ord('d') or k == ord('D'):
            print("지울 경로 :", rgb_dir)
            print("지울 경로 :", color_dir)
            print("지울 경로 :", depth_dir)
            
            f = open('deleted_images.txt', 'a')
            f.write("%s\n"%(rgb_dir))
            f.write("%s\n"%(color_dir))
            f.write("%s\n"%(depth_dir))
            f.write("%s\n"%('-------------------------'))
            os.remove(rgb_dir)
            os.remove(color_dir)
            os.remove(depth_dir)
            break
        elif k == 32:
                break
    cv2.destroyAllWindows()
    
home = './'

color_depth_dir = home + 'color_depth'
depth_dir = home + 'depth'
rgb_dir = home + 'rgb'

color_list = sorted(os.listdir(color_depth_dir))
depth_list = sorted(os.listdir(depth_dir))
rgb_list = sorted(os.listdir(rgb_dir))

n_color = len(color_list)
n_depth = len(depth_list)
n_rgb = len(rgb_list)

n = max(n_color, n_depth, n_rgb)

for i in range(n):
    rgb_file_dir = os.path.join(rgb_dir, rgb_list[i])
    color_file_dir = os.path.join(color_depth_dir, color_list[i])
    depth_file_dir = os.path.join(depth_dir, depth_list[i])
    
    print(rgb_file_dir)
    print(color_file_dir)
    print(depth_file_dir)
    print("")
    check(rgb_file_dir, color_file_dir,depth_file_dir)
    
    
```
