```python
import os
import shutil
file_path = './'
file_names = sorted(os.listdir(file_path))
name = input("name:")
save_path = './' + name
if not os.path.isdir(save_path):
    os.mkdir(save_path)
def change_name(input_txt,name):
    ub = '_'
    txt = input_txt.split(".")[0].split("_")
    if len(txt) == 3:
        txt = name + ub + txt[0] + ub + txt[1] + ub + txt[2] + '.txt' 
    elif len(txt) == 4:
        txt = name + ub + txt[0] + ub + txt[1] + ub + txt[2] + ub + txt[3] + '.txt' 
    return txt
```

```python
for old_name in file_names:
    if '.txt' not in old_name:
        pass
    else:
        new_name = change_name(old_name,name)
        print(new_name)
        src = os.path.join(file_path, old_name)
        dst = os.path.join(save_path, new_name)
        shutil.copy(src,dst)
```
