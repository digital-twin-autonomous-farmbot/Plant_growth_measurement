### 1. anotation 안된 파일 제거 / 2. flip 적용

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
import shutil

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print("Error : Creating Directory. " + directory)
        
deleted = './deleted'
createFolder(deleted)
d_cnt = 0

def imgg(x):
    if 'png' in str(x) or 'jpg' in str(x):
        return x
    else:
        return None

def txtt(x):
    if 'txt' in str(x):
        return x
    else:
        return None


home = './'

img_list = list(filter(imgg, sorted(os.listdir(home))))
txt_list = list(filter(txtt, sorted(os.listdir(home))))

n_img = len(img_list)
n_txt = len(txt_list)

while n_img != n_txt:
    img_list = list(filter(imgg, sorted(os.listdir(home))))
    txt_list = list(filter(txtt, sorted(os.listdir(home))))
    n_img = len(img_list)
    n_txt = len(txt_list)
    if n_img < n_txt:
        print("n_img < n_txt")
        for idx,txt in enumerate(txt_list):
            name = txt.split(".")[0] + '.png'
            if name not in img_list:
                txt_dir = os.path.join(home, txt_list[idx])
                delete_path = os.path.join(deleted, txt_list[idx])
                if not os.path.isfile(delete_path):
                    shutil.move(txt_dir, deleted)
        n_img = len(img_list)
        n_txt = len(txt_list)

    elif n_img > n_txt:
        print("n_img > n_txt")
        for idx,img in enumerate(img_list):
            name = img.split(".")[0] + '.txt'
            if name not in txt_list:
                img_dir = os.path.join(home, img_list[idx])
                delete_path = os.path.join(deleted, img_list[idx])
                if not os.path.isfile(delete_path):
                    shutil.move(img_dir, deleted)
        n_img = len(img_list)
        n_txt = len(txt_list)
            
if n_img == n_txt:
    print("n_img:{} == n_txt:{}".format(n_img, n_txt))
    n = n_img
    for idx in range(n):
        img_dir = os.path.join(home, img_list[idx])
        txt_dir = os.path.join(home, txt_list[idx])
        
        
        file_size = os.path.getsize(txt_dir)
        if file_size == 0:
            shutil.move(img_dir, deleted)
            shutil.move(txt_dir, deleted)
            print(txt_list[idx],"파일 크기 0이라 제거")
        else:
            # flip txt
            print("{}/{}".format(idx, n))
            lines = []
            f = open(txt_dir, 'r')
            while True:
                line = f.readline()
                if not line:
                    break
                lines.append(line.split(" "))
                
            f.close()
            
            flipped_txt = 'flipped_' + txt_list[idx]
            flipped_txt_dir = os.path.join(home, flipped_txt)
            
            f = open(flipped_txt_dir, 'w')
            for line in lines:
                
                line[1] = str(round(1-float(line[1]),6))
                line[2] = line[2]
                
                new_line = " ".join(line)
                f.write(new_line)
            f.close()
            
            #flip img
            flipped_img = cv2.flip(cv2.imread(img_dir),1)
            
            flipped_img_name = 'flipped_' + img_list[idx]
            flipped_img_dir = os.path.join(home, flipped_img_name)
            
            cv2.imwrite(flipped_img_dir, flipped_img)
            
```
