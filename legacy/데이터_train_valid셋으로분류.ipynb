```python
import os
import glob 
import shutil

folder_list = sorted(list(filter(os.path.isdir,os.listdir("./"))))
for idx,i in enumerate(folder_list):
    print("idx : {}, folder name : {}".format(idx, i))
n = int(input("\nprocess할 폴더 입력 :"))

target_folder = str(folder_list[n])
target_folder_dir = './' + target_folder

file_list = os.listdir(target_folder_dir)

file_txt = []
file_img = []

for file in file_list:
    if "txt" in file:
        file_txt.append(file)
    elif "png" in file:
        file_img.append(file)
        
file_txt = sorted(file_txt)
file_img = sorted(file_img)

n = len(file_img) // 10


idx_train = [(i+1) * 10 - 2  for i in range(n)]
idx_valid = [(i+1) * 5 - 1 for i in range(n*2)]


test_files_txt = []
test_files_img = []

valid_files_txt = []
valid_files_img = []

train_files_txt = []
train_files_img = []

for i in idx_train:
    test_files_txt.append(file_txt[i])
    test_files_img.append(file_img[i])
    
for i in idx_valid:
    valid_files_txt.append(file_txt[i])
    valid_files_img.append(file_img[i])
    
for i in range(len(file_txt)):
    if i in idx_train or i in idx_valid:
        pass
    else:
        train_files_txt.append(file_txt[i])
        train_files_img.append(file_img[i])
        
print("")
print("total Image :", len(file_img))
print("")
print("test txt :", len(test_files_txt))
print("valid txt:", len(valid_files_txt))
print("train txt:", len(train_files_txt))
print("")
print("test img :", len(test_files_img))
print("valid img:", len(valid_files_img))
print("train img:", len(train_files_img))
```
