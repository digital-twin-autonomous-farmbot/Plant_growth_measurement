```python
from email import header
from PIL import Image #image 파일 열기
import os #주소로 폴더 열기
import glob
import shutil
import csv
import pandas as pd
from pandas import read_csv

root_path = './'

test_dir = root_path + 'test'
train_dir = root_path + 'train'
valid_dir = root_path + 'valid'

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)
        
labels_dir = root_path + 'labels'

createFolder(labels_dir)

test_labels = os.listdir(test_dir)

cnt = 0
for test in test_labels:
    if 'txt' in test:
        if 'flipped' in test:
            pass
        else:
            file = test.split(".")[0].split("_")
            file.pop()
            file = "_".join(file) + '.txt'
            ori_dir = os.path.join(test_dir, test)
            file_dir = os.path.join(labels_dir,file)
            os.rename(ori_dir, file_dir)
            cnt += 1
    elif 'jpg' in test:
        if 'flipped' in test:
            pass
        else:
            file = test.split(".")[0].split("_")
            file.pop()
            file = "_".join(file) + '.jpg'
            ori_dir = os.path.join(test_dir, test)
            file_dir = os.path.join(test_dir,file)
            os.rename(ori_dir, file_dir)
            cnt += 1
        
print("{}개 label process done".format(cnt))
```
