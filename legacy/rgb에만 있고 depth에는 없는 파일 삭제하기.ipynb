### rgb에만 있고 depth, color depth에 없는 파일 삭제하기

```python
import os
import glob 
import shutil

images_folder_dir = './rgb'
csv_folder_dir = './depth'
        
images_file_list = sorted(os.listdir(images_folder_dir))
csv_file_list = sorted(os.listdir(csv_folder_dir))

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)
        
root_path = './'

deleted_labels_files = root_path + 'csv_deleted_files'

if not os.path.isdir(deleted_labels_files):
    createFolder(deleted_labels_files)
    
print(len(images_file_list))
print(len(csv_file_list))
```

```python
csv_list = os.listdir('./depth')
tmp = {}

for csv in csv_list:
    if 'flipped' in csv:
        tmp_name = csv.split("_")
        tmp_name = [tmp_name[0],'RGB',tmp_name[2],tmp_name[3].split(".")[0]]
        tmp_name = "_".join(tmp_name)
        tmp[csv] = tmp_name + '.jpg'
    else:
        tmp_name = csv.split("_")
        tmp_name = ['RGB',tmp_name[1],tmp_name[2].split(".")[0]]
        tmp_name = "_".join(tmp_name)
        tmp[csv] = tmp_name + '.jpg'
        
csv_dict = tmp
```

```python
csv_dict_keys = list(csv_dict.keys())
csv_dict_values = list(csv_dict.values())
for image in images_file_list:
    if image in csv_dict_values:
        idx = csv_dict_values.index(image)
        depth = csv_dict_keys[idx]
        src = os.path.join('./depth', depth)
        dst = os.path.join('./tmp', depth)
        os.rename(src, dst)
```
